```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales-of-shampoo-over-a-three-ye.csv')
```

```python
df.shape
```

```python
df.columns = ['Months','price']
```

```python
df.head()
```

```python
plt.plot(df.price)
```

```python
# dicky-fuller test
from statsmodels.tsa.stattools import adfuller
adfuller(df.price)
```

```python
adfuller(df.price.diff(periods=1).dropna())
```

```python
plt.plot(df.price.diff(periods=1).dropna())
```

```python
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
plot_pacf(df.price.diff(periods=1).dropna());
# p-6
```

```python
plot_acf(df.price.diff(periods=1).dropna());
# q-1
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df.price,order=(6,1,1))
result = model.fit(disp=0)
```

```python
result.plot_predict(start=1,end=50);
```
